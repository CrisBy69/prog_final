    <h2 style="color: darkred;">1. INTRODUCCIÓN</h2>
</div>

En las últimas décadas, las ciudades han experimentado un crecimiento sostenido en su población y en la actividad urbana, lo que ha planteado importantes desafíos en términos de movilidad, sostenibilidad ambiental y calidad de vida. Frente a problemas como la congestión del tráfico, la contaminación atmosférica y la creciente demanda de alternativas de transporte eficientes, se ha vuelto imprescindible transformar los modelos tradicionales de movilidad hacia sistemas más sostenibles, accesibles y respetuosos con el medio ambiente.

En este contexto, el uso de la bicicleta como medio de transporte urbano ha cobrado relevancia, aunque su desarrollo ha atravesado diversas etapas y muchas no lograron consolidarse plenamente. Este fenómeno se explica, en buena medida, por la carencia de infraestructuras adecuadas y por el avance de otros sistemas de transporte más profesionalizados, como las redes de autobuses y metro, que ofrecen mayor comodidad y accesibilidad. Paralelamente, la expansión del vehículo privado, impulsada por políticas económicas y comerciales, relegó a la bicicleta a un papel secundario, percibiéndose mayormente como un medio de ocio más que como una alternativa de movilidad funcional (Mirea, 2017).

Para revertir esta situación y fomentar el uso diario de la bicicleta, a comienzos del siglo XXI varios gobiernos europeos promovieron la implementación de sistemas públicos de alquiler de bicicletas. Estas iniciativas han contribuido a integrar la bicicleta dentro del modelo de transporte urbano, garantizando la seguridad y la convivencia con otros medios (Mirea, 2017). Generalmente, estos sistemas permiten a los usuarios acceder a las bicicletas mediante una tarjeta o aplicación móvil, abonando una tarifa accesible, habitualmente en forma de suscripción anual o mensual. Esta facilidad de acceso ha sido clave para consolidar la bicicleta como un elemento fundamental en la movilidad urbana sostenible.

Este impulso se reflejó en Madrid en 2014 con la creación de BiciMAD, el sistema público de bicicletas. Aunque el sistema de transporte público madrileño es considerado uno de los mejores a nivel mundial, el principal reto de futuro radica en promover el cambio de hábitos de los viajeros, no solo desde el automóvil privado hacia el transporte público colectivo (metro, cercanías, autobuses), sino también hacia modos de transporte no motorizados como la caminata y la bicicleta (Carballo, 2014).

BiciMAD tuvo que enfrentar desafíos específicos derivados de la orografía de Madrid y de una red ciclista aún limitada en comparación con otras capitales europeas. Dos años más tarde de su creación, la gestión del servicio fue asumida directamente por el Ayuntamiento, con el objetivo de consolidar BiciMAD como una alternativa real y eficiente para el transporte urbano. Desde entonces, se han impulsado diversas iniciativas para mejorar la infraestructura ciclista, aumentar la disponibilidad de bicicletas y estaciones y fomentar su uso cotidiano entre los ciudadanos (Mirea, 2017).

Uno de los aspectos diferenciadores de BiciMAD es su carácter eléctrico. Esto supone una ventaja significativa al hacer más atractiva la bicicleta para aquellos usuarios no habituales ya que facilita en rapidez y maniobralidad en el tránsito urbano. Sin embargo, la ausencia de una red consolidada de carriles bici segregados del tráfico motorizado, junto con las condiciones del sistema tarifario, constituyen factores que podrían limitar la expansión y aceptación del servicio entre los usuarios potenciales (Carballo, 2014).

No obstante, para que este tipo de servicios funcione de manera eficiente, es fundamental comprender cómo y cuándo se utilizan las bicicletas con el fin de ajustar la oferta a la demanda real. Comprender los patrones de uso de BiciMAD permite optimizar la distribución de bicicletas y estaciones y facilita la toma de decisiones estratégicas para mejorar la experiencia del usuario. Por lo tanto, la predicción de la demanda se ha convertido en una herramienta fundamental para los gestores del servicio. De esta manera, es posible anticiparse a picos de utilización, prevenir la escasez o exceso de bicicletas en determinadas zonas y adaptar los recursos a las condiciones reales de la ciudad.

La demanda de bicicletas puede verse afectada por múltiples factores, entre ellos las condiciones meteorológicas, la hora del día, el día de la semana, la estacionalidad, eventos especiales o cambios en la infraestructura urbana. Analizar e integrar estas variables en modelos predictivos permite obtener estimaciones más precisas sobre el comportamiento futuro de los usuarios y, por tanto, mejorar la planificación del servicio.

Este trabajo tiene como objetivo principal desarrollar un modelo de predicción del uso del sistema BiciMAD a partir del análisis de datos históricos principalmente. Para ello, se emplearán técnicas de análisis de datos y aprendizaje automático (machine learning), que permiten identificar patrones complejos y realizar predicciones con altos niveles de precisión. A través de este estudio, se pretende mejorar la gestión del sistema de bicicletas compartidas en Madrid y contribuir al desarrollo de herramientas que apoyen la planificación urbana sostenible y orientada a datos.

#### 3. Factores condicionantes

El uso de la bicicleta no sigue un patrón lineal en el tiempo, estas preferencias de uso responden a una serie -en mayor medida- de factores ambientales y estacionales. A continuación, se detallan los principales factores ambientales identificados en la literatura existente con el fin de proporcionar una base sólida para el diseño de los modelos predictivos teniendo en consideración la complejidad del fenómeno.

- **Tipo de día**
El tipo de día (laborable o fin de semana) influye directamente en los patrones de movilidad de la población. Durante los días laborables, el uso de la bicicleta suele estar vinculado a desplazamientos obligados -centro de trabajo o estudios-, reflejándose un mayor uso en hora punta. En cambio, los fines de semana y festivos suelen manifestar un uso más recreativo con horarios más dispersos y recorridos distintos, generalmente menos dirigidos al centro urbano. 

- **Condiciones metereológicas**
Las condiciones meteorológicas influyen tanto en el uso de la bicicleta como la frecuencia de uso. Ya no solo por los efectos estacionales a largo plazo, sino, además, porque las inclemencias climáticas pueden cambiar constantemente en plazos muy breves (Eren & Uz, 2020). De entre todas, destaca temperatura y precipitaciones.

-	*Temperatura*
La temperatura posee gran influencia especialmente en contextos urbanos y de movilidad diaria, el objeto de estudio, debido al riesgo que representa para la salud, particularmente en climas mediterráneo-continentalizados como el de Madrid. Diversos estudios han demostrado una correlación positiva entre el aumento temperatura y la demanda de bicicletas. La generación de viajes tiende a incrementarse cuando las temperaturas oscilan entre los 0 y 20 °C, alcanzando su nivel más alto en el rango de 20 a 30 °C. Sin embargo, los efectos observados a partir de temperaturas superiores a los 30 °C presentan resultados poco concluyentes. (Eren & Uz, 2020).

-	*Precipitaciones*
Se considera uno de los factores determinantes, ya que las precipitaciones y las bajas temperaturas representan las condiciones más desfavorables. De acuerdo con Eren y Uz (2020), se requieren aproximadamente tres horas para que la demanda retome su nivel habitual tras estas situaciones.

- **Estación del año**

En contraposición al análisis exhaustivo de los movimientos registrados entre 2017 y 2018, llevado a cabo por Martín Sánchez (2019), que confirma que la demanda de BiciMAD aumenta durante los meses de primavera y verano -alcanzando su punto máximo entre junio y julio- y disminuye progresivamente el número de desplazamientos durante otoño e invierno, los resultados obtenidos en el Análisis Exploratorio de Datos (EDA) para el año 2022 muestran un comportamiento diferente. Aunque marzo presenta las mayores tasas de uso, se observa una mayor constancia en la demanda durante los meses de otoño e invierno, sugiriendo una influencia estacional distinta en este período. Estos hallazgos evidencian que la estacionalidad sigue siendo un factor determinante en la demanda del servicio, aunque su patrón puede variar entre diferentes años o condiciones específicas.

Teniendo en cuenta estos factores, se consideró oportuno ponderarlos durante el proceso de entrenamiento del modelo con el fin de optimizar su rendimiento y generar predicciones más ajustadas a contextos reales. Cabe destacar que dicha ponderación ha sido estimada de forma empírica, sin basarse en estudios previos específicos, sino en supuestos razonables derivados del conocimiento general sobre el impacto en el uso de la bicicleta.

```python
Image('modelos.jpg')
```

Como se puede apreciar, XGBoost (Extreme Gradient Boosting) obtuvo el menor MSE y el mayor R². Esto demuestra que, a la vez que minimiza el error de predicción, también explica de manera más completa la variabilidad del uso del sistema. Este algoritmo, basado en el ensamblaje de árboles de decisión mediante un proceso de boosting, es especialmente útil para capturar relaciones no lineales y complejas en los datos, además de incluir regularización para evitar el sobreajuste. 

En cuanto a Random Forest que, a diferencia del primero, este algoritmo utiliza ensamblaje paralelo de múltiples árboles entrenados con muestras aleatorias del conjunto de datos. Este hecho ayuda a reducir la varianza y mejorar la robustez del modelo frente al ruido. Sin embargo, aunque suele ser una técnica muy acertada en muchas situaciones, en este caso su MSE fue mayor y su R² menor que los de XGBoost, demostrando que capturó menos detalle en las relaciones complejas del conjunto de datos.

Por último, a pesar de que el método de Redes Neuronales (Neural Network), inspirado en el funcionamiento del cerebro humano, tiene gran potencial para modelar datos complejos, requieren más datos y ajustes para alcanzar un rendimiento óptimo. En este caso, mostró un desempeño inferior debido a la naturaleza y tamaño del conjunto de datos utilizado.

La superioridad de XGBoost queda reflejada en ambas métricas por su capacidad para combinar flexibilidad, manejo eficiente de interacciones complejas y control del sobreajuste.

El desarrollo de una aplicación capaz de predecir el uso de bicicletas en un servicio de transporte público como BiciMAD representa un avance significativo en la gestión inteligente de la movilidad urbana. Este tipo de innovación responde tanto a las necesidades crecientes de una población cada vez más digitalizada y que, a la par, se alinea con los Objetivos de Desarrollo Sostenible (ODS) propuestos por la ONU, en especial con aquellos relacionados con el desarrollo de ciudades sostenibles, la acción por el clima y la mejora en la eficiencia de los sistemas de transporte.

En la actualidad, los usuarios de BiciMAD tienen acceso, a través de la propia plataforma del servicio, a información en tiempo real sobre la disponibilidad de bicicletas y anclajes en cada estación. Ahora bien, a pesar de que supone una mejora, también presenta limitaciones importantes ya que no permite planificar con antelación ni anticiparse a los picos de demanda que pueden dejar una estación sin bicicletas o sin espacio para aparcarlas.

A raíz de ello, la nueva aplicación propuesta va un paso más allá. Gracias al uso de técnicas avanzadas de análisis de datos y aprendizaje automático (machine learning), se ha podido entrenar un modelo predictivo utilizando datos históricos de uso del sistema. A partir de esta información, el sistema es capaz de anticipar la demanda futura en cada estación, estimando con una alta fiabilidad cuántas bicicletas estarán disponibles en un momento concreto del día. Esta aplicación abre un amplio abanico de posibilidades para los usuarios: personas que dependen de la bicicleta pública para sus desplazamientos diarios podrán planificar sus rutas con antelación y asegurarse de contar con una bicicleta cuando la necesiten. Además, también mejora la experiencia de usuarios ocasionales que podrán planificarse sin temor a contratiempos.

Desde el punto de vista operativo, este modelo predictivo también supone una ventaja para la propia gestión del sistema ya que permite a los operadores del servicio optimizar la redistribución de bicicletas entre estaciones, anticiparse a los desequilibrios y mejorar el propio servicio. 

No obstante, durante el desarrollo del modelo se han identificado una serie de limitaciones que podrían convertirse en oportunidades de mejora para futuras líneas de trabajo.

Por ejemplo, habría sido deseable poder predecir la demanda de bicicletas en cualquier intervalo horario a lo largo del día y no únicamente entre las 14:00 y las 15:00 horas. Aunque esta franja resulta significativa por coincidir con un periodo de alta actividad laboral y académica, restringe la versatilidad del modelo al no permitir consultas en otras horas, limitando su aplicabilidad práctica. Dado que se dispone de los datos necesarios para extender el modelo a todo el día, se considera recomendable abordar esta mejora en futuros desarrollos. En este caso, la decisión de no hacerlo responde a la necesidad de acotar el alcance del presente trabajo. Asimismo, otra mejora a contemplar sería la incorporación de la variable "día festivo", tomando como referencia el calendario oficial de cada año. Este tipo de días suele presentar patrones de uso significativamente distintos respecto a los laborables o fines de semana, concediéndole al modelo una mayor precisión y realismo. El hecho de incluir esta distinción permitiría registrar mejor la variabilidad de la demanda en fechas atípicas, que a menudo escapan a las tendencias regulares.

Otra limitación detectada fue la relacionada con el número de estaciones consideradas. Al extraer los datos de uso, se observó que solo estaban presentes 307 estaciones, cuando según la propia web de la [EMT](https://www.emtmadrid.es/Noticias/Continua-el-crecimiento-de-bicimad-el-sistema-se.aspx), deberían figurar unas 650. Esto sugiere que el modelo ha sido entrenado sobre una muestra parcial y, por ende, generando un sesgo y reduciendo la representatividad de los resultados respecto a la red completa de BiciMAD.

Por último, también se reconoce como una oportunidad de mejora el apoyo en literatura científica y estudios previos para establecer de forma más sólida el impacto de las variables en el comportamiento de la demanda e, incluso, añadir algunas más. 

A pesar de estas limitaciones, esta propuesta constituye un claro ejemplo del potencial que tiene el análisis de datos para contribuir activamente a la construcción de sistemas de movilidad urbana más inteligentes y adaptativos. En un escenario de cambio constante, soluciones de este tipo permiten avanzar para marcar la diferencia entre un servicio útil y uno verdaderamente inteligente.
